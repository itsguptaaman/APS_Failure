### Droping the columns having missing values more than 50%

```python
miss=df.isna().mean()*100 < 50
```

```python
columns=miss[miss==True].index
```

```python
df=df[columns].copy()
```

```python
df.sample(5)
```

```python
df.shape
```

```python
# filling the nan values
```

```python
df["aa_000"].var()
```

```python
columns=columns[1:]
```

```python
variance=[np.round(df[i].var()) for i in columns ]
```

```python
pd.DataFrame(variance, columns)
```

```python
def Random_Sample_imputation(feature,df):
    random_sample=df[feature].dropna().sample(df[feature].isnull().sum())               
    random_sample.index=df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature]=random_sample
```

```python
for col in columns:
    Random_Sample_imputation(col,df)
```

```python
df.isna().sum()
```

```python
variance=[np.round(df[i].var()) for i in columns ]
```

```python
pd.DataFrame(variance, columns)
```

### After random imputation there is no change in variance
